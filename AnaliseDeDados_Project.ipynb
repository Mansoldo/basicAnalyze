<h3>Passo 1 - Trazer sua base de dados para o Python e ver o que tem nela

```python
import pandas as pd

tabela = pd.read_excel("/home/Vendas.xlsx")
display(tabela)
```

<h3>Passo 2 - Pegar um panorama geral sobre a sua base de dados</h3>

```python
faturamento_total = tabela['Valor Final'].sum()
print(f'Faturamento total de: {faturamento_total}')
```

<h3>Passo 3 - Começar análise Top -> Down

```python
#Faturamento por loja
faturamento_por_loja = tabela[['ID Loja','Valor Final']].groupby('ID Loja').sum('Valor Final')
display(faturamento_por_loja)
```

<h3>Passo 4 - Entrar no detalhe para entender

```python
#Faturamento por produto
faturamento_por_produto = tabela[['ID Loja','Produto','Valor Final']].groupby(['ID Loja','Produto']).sum()
display(faturamento_por_produto)
```
